```python
a= sns.countplot(data=df,x='transmission')
a.set_title('Transmission count')
a.set_xlabel('Transmission Values')
a.set_ylabel('Transmission count')
a.set_xticklabels(['Ma','Auto'])
```

```python
a= sns.countplot(data=df,x='transmission', order=['Automatic','Manual'], width=0.3)
```

```python
#hue 
```

```python
a= sns.countplot(data=df,x='transmission',hue='fuel_type', order=['Automatic','Manual'])
```
